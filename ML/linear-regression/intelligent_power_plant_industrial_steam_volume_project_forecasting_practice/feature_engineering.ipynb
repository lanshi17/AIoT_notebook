# 归一化

```python
#最值归一化
columns=list(all_data.columns)
columns.remove('label')
columns.remove('target')
all_data[columns].describe()
```

```python
def norm_min_max(col):
    """
    对一列数据进行最小-最大规范化处理。

    参数:
    col (pd.Series): 需要进行规范化处理的数据列。

    返回:
    pd.Series: 规范化处理后的数据列。
    """
    # 计算数据列的最小值和最大值，并用以规范化数据列
    return (col-col.min())/(col.max()-col.min())

# 对指定的列进行最小-最大规范化处理
all_data_normed = all_data[columns].apply(norm_min_max, axis=0)

# 显示规范化后数据的基本统计信息
all_data_normed.describe()
```

```python
# 实例化MinMaxScaler，用于后续的数据归一化处理
min_max_scaler = preprocessing.MinMaxScaler()

# 使用MinMaxScaler对选定的列进行数据归一化处理
# fit_transform方法既拟合数据，又转换数据
all_data_normed = min_max_scaler.fit_transform(all_data[columns])

# 将归一化后的数据转换为DataFrame格式，便于后续的数据操作和分析
all_data_normed = pd.DataFrame(all_data_normed, columns=columns)

# 显示归一化后数据的基本统计信息，如均值、标准差等
all_data_normed.describe()
```

# 定义数据处理函数

```python
# 导入数据分割工具
from sklearn.model_selection import train_test_split

# 获取训练数据
# 该函数用于从全局数据集中提取标记为"train"的数据，并分离特征和目标变量
def get_train_data():
    # 筛选训练数据
    train_data=all_data[all_data["label"]=="train"]
    # 提取目标变量
    y=train_data.target
    # 移除目标变量和标签，保留特征
    X=train_data.drop(["target","label"],axis=1)
    # 返回特征和目标变量
    return X,y

# 分割训练数据
# 该函数用于将标记为"train"的数据集分割为训练集和验证集
def split_train_data():
    # 筛选训练数据
    train_data=all_data[all_data["label"]=="train"]
    # 提取目标变量
    y=train_data.target
    # 移除目标变量和标签，保留特征
    X=train_data.drop(["target","label"],axis=1)
    # 分割数据集，20%作为验证集
    X_train,X_valid,y_train,y_valid=train_test_split(X,y,test_size=0.2)
    # 返回训练集和验证集的特征及目标变量
    return X_train,X_valid,y_train,y_valid

# 获取测试数据
# 该函数用于从全局数据集中提取标记为"test"的数据，并移除不必要的列
def get_test_data():
    # 筛选测试数据，并重置索引
    test_data=all_data[all_data["label"]=="test"].reset_index(drop=True)
    # 移除标签和目标变量，返回测试数据的特征
    return test_data.drop(["label","target"],axis=1)
```

# 异常值过滤


```python
# 异常值过滤
from sklearn.metrics import mean_squared_error
def find_outliers(model,X,y,sigma=3):
    model.fit(X,y)
    y_pred=pd.Series(model.predict(X),index=y.index)

    resid=y-y_pred
    mean_resid=resid.mean()
    std_resid=resid.std()

    Z=(resid-mean_resid)/std_resid
    outliers=Z[abs(Z)>sigma].index

    print("R^2=",model.score(X,y))
    print("MSE=",mean_squared_error(y,y_pred))
    print("-----------------------------------")

    print('mean of residuals:',mean_resid)
    print('std of residuals:',std_resid)
    print('---------------------------------------')

    print("outliers:",len(outliers))
    print(outliers.tolist())

    # 真实值预测值散点图
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1)
    plt.plot(y,y_pred,".")
    plt.plot(y.loc[outliers],y_pred.loc[outliers],"ro")
    plt.legend(['Accepted','Outlier'])
    plt.xlabel('y')
    plt.ylabel('y_pred');

    # 真实值和残差散点图
    plt.subplot(1,3,2)
    plt.plot(y,resid,".")
    plt.plot(y.loc[outliers],resid.loc[outliers],"ro")
    plt.legend(['Accepted','Outlier'])
    plt.xlabel('y')
    plt.ylabel('y - y_pred');
    # 直方图
    ax_133=plt.subplot(1,3,3)
    Z.plot.hist(bins=50,ax=ax_133)
    Z.loc[outliers].plot.hist(color='r',bins=50,ax=ax_133)
    plt.legend(['Accepted','Outlier'])
    plt.xlabel('Z')

    plt.savefig("./export_data/outliers.png")

    return outliers
```

## 套索回归算法检测异常值

$
    J(\mathbf{w}) = \frac{1}{2} ||\mathbf{y} - X\mathbf{w}||_2^2 + \alpha ||\mathbf{w}||_1 
$

其中，$\mathbf{y}$ 是目标变量，$X$ 是特征矩阵，$\mathbf{w}$是权重向量，$\alpha$ 是正则化强度参数，$||\cdot||_1 $表示L1范数[1]。


```python
from sklearn.linear_model import Lasso
X_train,y_train=get_train_data()
outliers2=find_outliers(Lasso(),X_train,y_train)
```

## 支持向量机SVR算法检测异常值
支持向量机（Support Vector Machine, SVM）是一种用于分类和回归任务的监督学习算法。支持向量机的核心思想是找到一个能够最好地分离数据的超平面，同时允许一定的误差范围。支持向量机回归（Support Vector Regression, SVR）是SVM在回归任务中的应用版本。
在SVR中，我们试图找到一个函数 \( f(x) \)，使得对于每一个训练样本 \( (x_i, y_i) \)，都有 \( |y_i - f(x_i)| \leq \epsilon \)，其中 \(\epsilon\) 是一个预先设定的误差容限。为了实现这一目标，SVR使用了\(\epsilon\)-不敏感损失函数，这意味着只有当预测值与实际值之间的差异超过\(\epsilon\)时，才会对损失函数做出贡献。
SVR模型通过引入拉格朗日乘子和核函数来解决上述优化问题。核函数的选择对于SVR的表现至关重要，因为它决定了数据在高维空间中的映射方式。常见的核函数包括线性核、多项式核、径向基函数（RBF）核等。

```python
# 获取训练数据
from sklearn.svm import SVR
X_train, y_train = get_train_data()

# 使用岭回归查找异常值
outliers3 = find_outliers(SVR(), X_train, y_train)
```

## Xgboost算法检测异常值
XGBoost的优化目标是通过最小化以下目标函数来实现的：

 $$L(\theta) = \sum_{i=1}^{n} l(y_i, \hat{y}_i) + \sum_{k=1}^{K} \Omega(f_k)$$ 

其中，\( $l(y_i, \hat{y}_i)$ \) 是样本 $i$ 的损失函数，\( $\Omega(f_k)$ \) 是树 \( $k$ \) 的正则化项，\( $K$ \) 是树的数量。

```python
# 获取训练数据
from xgboost import XGBRegressor
X_train, y_train = get_train_data()

# 使用岭回归查找异常值
outliers4 = find_outliers(XGBRegressor(), X_train, y_train)
```

# 多重共线性

## 多重共线性
是指各特征之间存在线性相关关系，即一个特征可以是其他一个或几个特征的线性组合。如果存在多重共线性，求损失函数时矩阵会不可逆，导致求出结果会与实际不同，有所偏差。

说白了就是多个特征存在线性关系，数据冗余，但不完全是，所以要将成线性关系的特征进行降维。

用线代的知识来说就是奇异矩阵.
